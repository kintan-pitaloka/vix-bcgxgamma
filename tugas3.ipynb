```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA

# Data runtun waktu
data = [
    list(range(1, 11)),
    list(range(11, 21)),
    list(range(21, 31)),
    list(range(31, 41)),
    list(range(41, 51)),
    list(range(51, 61)),
    [102], [99], [101], [97], [102], [100], [101], [96], [105], [99],
    [100], [96], [104], [100], [95], [100], [99], [105], [100], [96],
    [100], [94], [100], [103], [100], [99], [102], [98], [100], [99],
    [103], [98], [100], [103], [97], [104], [96], [104], [99], [105],
    [97], [102], [103], [98], [101], [98], [100], [103], [102], [94],
    [105], [96], [103], [100], [103], [98], [100], [97], [101], [98]
]

# Mengubah daftar dalam daftar menjadi array 1D
flattened_data = np.concatenate(data)

# Plot data asli
plt.plot(flattened_data)
plt.title('Data Runtun Waktu')
plt.xlabel('Indeks')
plt.ylabel('Nilai')
plt.show()

# Uji ADF untuk stasionaritas
result = adfuller(flattened_data)
print('ADF Statistic:', result[0])
print('p-value:', result[1])

if result[1] >= 0.05:
    # Jika data tidak stasioner, lakukan differencing
    diff_data = np.diff(flattened_data)
    plt.plot(diff_data)
    plt.title('Data Setelah Differencing')
    plt.xlabel('Indeks')
    plt.ylabel('Nilai')
    plt.show()
    
    # Uji ADF pada data yang sudah di-difference
    result_diff = adfuller(diff_data)
    print('ADF Statistic after differencing:', result_diff[0])
    print('p-value after differencing:', result_diff[1])
    
    data_to_use = diff_data
else:
    data_to_use = flattened_data

# Plot ACF dan PACF
plt.figure(figsize=(12, 6))

plt.subplot(121)
plot_acf(data_to_use, lags=20, ax=plt.gca())
plt.title('ACF')

plt.subplot(122)
plot_pacf(data_to_use, lags=20, ax=plt.gca())
plt.title('PACF')

plt.tight_layout()
plt.show()

# Identifikasi nilai p dan q dari plot ACF dan PACF, misalkan ARIMA(1, 1, 1)
p = 1
d = 1 if result[1] >= 0.05 else 0
q = 1

# Pasang model ARIMA
model = ARIMA(flattened_data, order=(p, d, q))
model_fit = model.fit()

print(model_fit.summary())
```
